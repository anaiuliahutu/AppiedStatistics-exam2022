```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
def count_zero(x):
    zero=0
    for i in x:
        if i==0:
            zero=zero+1
    return zero 
```

```python
def count_one(x):
    one=0
    for i in x:
        if i==1:
            one=one+1
    return one 
```

```python
dataframe = pd.read_csv('data_AnorocDisease.csv', header=None)
df=dataframe[0].str.split(';', expand=True)
df.columns=df.iloc[0]
df=df[1:]
ID, temp, bloodp, age, status=df.values.T.astype(float)
control_sample=df[0:800].astype(float)
unknown_sample=df[800:1000]
ID, temp, bloodp, age, status=control_sample.values.T.astype(float)
```

```python
#Problem 4.1.1
age = ['   Age','      Status (0: Healthy']
control_sample.groupby(age).size().unstack().plot(kind='kde', stacked=True, title='Age by status')
plt.legend(['Healthy', 'Sick'])
means=control_sample.groupby(age).size().unstack().mean()
std=control_sample.groupby(age).size().unstack().std()
n=count_zero(control_sample['      Status (0: Healthy'])
print(means)
print(std)
print(n)
print(800-n)
t_test_age=(means[0]-means[1])/math.sqrt(std[0]**2/n+std[1]**2/(800-n))
print("T-test for age is: "+ str(t_test_age))
```

```python
control_sample.groupby(['   BloodP','      Status (0: Healthy']).size().unstack().plot(kind='kde', stacked=True, title='Blood pressure by status')
plt.legend(['Healthy', 'Sick'])
means=control_sample.groupby(['   BloodP','      Status (0: Healthy']).size().unstack().mean()
std=control_sample.groupby(['   BloodP','      Status (0: Healthy']).size().unstack().std()
n=count_zero(control_sample['      Status (0: Healthy'])
print(means)
print(std)
print(n)
t_test_blood=(means[0]-means[1])/math.sqrt(std[0]**2/n+std[1]**2/(800-n))
print("T-test for blood pressure is: "+ str(t_test_blood))
```

```python
control_sample.groupby(['  Temp','      Status (0: Healthy']).size().unstack().plot(kind='kde', stacked=True, title='Temperature by status')
plt.legend(['Healthy', 'Sick'])
means=control_sample.groupby(['  Temp','      Status (0: Healthy']).size().unstack().mean()
std=control_sample.groupby(['  Temp','      Status (0: Healthy']).size().unstack().std()
n=count_zero(control_sample['      Status (0: Healthy'])
print(means)
print(std)
print(n)
t_test_temp=(means[0]-means[1])/math.sqrt(std[0]**2/n+std[1]**2/(800-n))
print("T-test for temperature: "+ str(t_test_temp))
```
