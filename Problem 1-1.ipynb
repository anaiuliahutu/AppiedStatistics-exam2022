```python
import math
import numpy as np
from scipy.stats import norm, multivariate_normal
from matplotlib import pyplot as plt
```

```python
# Problem 1.1.1

mean=50
std=20
a=55
b=65
x=np.arange(0,100,1)
y=np.arange(0,100,1)
#Plot
plt.plot(x, norm.pdf(x,mean,std))
plt.xlabel('Scores in test A')
plt.ylabel('Frequency')
plt.title('Gaussian Distribution')
#Interval delimitation
plt.fill_between(x, norm.pdf(x,mean,std), where=(x<=65)&(x>=55), color='g')
plt.show()
Fraction=norm.cdf((b-mean)/std)-norm.cdf((a-mean)/std)
print("The fraction is: "+str(Fraction))
```
